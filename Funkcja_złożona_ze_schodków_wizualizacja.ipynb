```python
import numpy as np
import matplotlib.pyplot as plt
```

# Przybliżenie funkcji przez schodki - wizualizacja

```python
def step_approximation(f, x_range, num_neurons):
    """
    Makes a visualisation of the step approximation of a function.
    Plots the function and the approximation.
    Each step is a plat

    Parameters:
    - f - function to approximate
    - x_range - tuple of the form (x_min, x_max) that defines the range of x values
    - num_steps - number of steps to use
    """
    assert num_neurons >= 1, "num_neurons must be greater than 1"
    dx = (x_range[1] - x_range[0]) / (num_neurons + 1)
    x_values = [x_range[0],]
    y_values = [f(x_range[0]),]
    for i in range(num_neurons):
        # Append a step
        x_values.append(x_values[-1] + dx)
        x_values.append(x_values[-1])
        y_values.append(y_values[-1])
        y_values.append(f(x_values[-1]))
    # Final value
    x_values.append(x_range[1])
    y_values.append(y_values[-1])
    # Create list with ground truth to plot
    x_s = np.linspace(x_range[0], x_range[1], 100)
    y_s = [f(x) for x in x_s]
    # Plot the function and the approximation
    plt.plot(x_values, y_values, label = "Przybliżenie")
    plt.plot(x_s, y_s, label = "Funkcja", alpha = 0.5)
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
```

Funkcja:
$$f(x) = \frac{(x-3)^2 (x+3)^3 + 10x}{100}$$

Przedział:
$$-5<x<5$$

```python
NUMBER_OF_STEPS = 100

def f(x):
    return ((x-3)**2 * (x+3)**2 + 10*x) / 100

step_approximation(
    f,
    (-5, 5),
    num_neurons = NUMBER_OF_STEPS)
```
